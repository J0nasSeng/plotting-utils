# Plotting with Standard Deviation
Often, we have e.g. accuracy values of multiple experiments we want to plot. In such cases it is often beneficial to see the mean-accuracy for each epoch of training and the standard deviation of the accuracy-values per epoch. The following code allows us to do visualize such data.

```python
acc = np.array([0.1, 0.6, 0.7, 0.8, 0.81])
accs = np.array([acc + np.random.normal(0, 0.02, len(acc)) for _ in range(0, 10)])
mean_accs = accs.mean(axis=0)
lowers = mean_accs - mean_accs.std(axis=0)
uppers = mean_accs + mean_accs.std(axis=0)
```

```python
fig, ax = plt.subplots(figsize=(12, 9))
x = np.arange(0, len(mean_accs))
ax.tick_params(which='major', size=10, width=2, direction='in', right='on', top='on')
ax.tick_params(which='minor', size=7, width=2, direction='in', right='on', top='on')
plt.plot(x, mean_accs, marker='x', label='marker')
plt.fill_between(x, lowers, uppers, alpha=.3)

plt.title('This is a lineplot with standard deviations')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.legend(loc='lower right')
plt.show()
```

# Regression Plot with Confidence Interval
Another popular plot which depicts simlar data are regression plots. Here, we fit a linear regression model to the data passed and plot a x%-confidence interval around the fit. This can be done using the following lines.

```python
x = np.random.normal(0, 1, 100)
y = 2*x + np.random.normal(0, 1, 100)
```

```python
fig, ax = plt.subplots(1, 1, figsize=(12, 9))
ax.tick_params(which='major', size=10, width=2, direction='in', right='on', top='on')
ax.tick_params(which='minor', size=7, width=2, direction='in', right='on', top='on')
ax.scatter(x, y, color='red', label='Scatter points')
sns.regplot(x=x, y=y, scatter_kws={'zorder': -1}, ax=ax, color='blue')
plt.title('This is a regplot')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.legend(loc='lower left')
```
